# Veri Hazırlama

```python
from sklearn.preprocessing import MinMaxScaler
import pickle
from tqdm.notebook import tnrange
```

```python
# yalnızca gerekli verileri filtrele
data = data[['Close', 'Volume']]
data.head(5)
```

```python
# test seti uzunluğunu onaylayın
test_length = data[(data.index >= '2022-10-01')].shape[0]
```

Daha sonra bir veri çerçevesi ve özellik uzunluğunu alan ve özellikleri ve hedefleri döndüren bir fonksiyon tanımladık.

```python
def CreateFeatures_and_Targets(data, feature_length):
    X = []
    Y = []
    
    for i in tnrange(len(data) - feature_length):
        X.append(data.iloc[i : i+feature_length,:].values)
        Y.append(data["Close"].values[i+feature_length])
    
    X = np.array(X)
    Y = np.array(Y)
    
    return X, Y
```

İlk argüman (veri), 'Kapat' ve 'Hacim' sütunlarına sahip verilerimizdir.

İkinci argüman (özellik_uzunluğu) geçmiş veri noktalarının sayısıdır.

```python
X, Y = CreateFeatures_and_Targets(data, 32)
```

```python
# Şekilleri kontrol edin
X.shape, Y.shape
```

X üzerinde 3 boyutlu verimiz, Y üzerinde ise 1 boyutlu verimiz var, **ilk** boyut özelliklerimizdeki kayıt sayısıdır. **İkincisi** seçilen özellik uzunluğudur ve **üçüncüsü** filtrelenen verilerdeki sütun sayısıdır.

Daha sonra özelliklerimizi ve hedefimizi tren testi bölümüyle bölüyoruz.

```python
Xtrain, Xtest, Ytrain, Ytest = X[:-test_length], X[-test_length:], Y[:-test_length], Y[-test_length:]
```

```python
# Eğitim veri kümesi şeklini kontrol edin
Xtrain.shape, Ytrain.shape
```

```python
# Test veri kümesi şeklini kontrol edin
Xtest.shape, Ytest.shape
```

3 boyut özelliğimiz olduğundan ve ölçekleyici yalnızca 2 boyutta ölçeklendirmeyi desteklemektedir. Böylece bir vektörü birden çok boyutla ölçeklendirecek bir fonksiyon yaratıyoruz.

```python
# vektörleri birden çok boyutla ölçeklendirmek için bir ölçekleyici oluşturun
class MultidimensionScaler():
    def __init__(self):
        self.scalers = []
        
    def fit_transform(self, X):
        total_dims = X.shape[2]
        for i in range(total_dims):
            Scaler = MinMaxScaler()
            X[:, :, i] = Scaler.fit_transform(X[:, :, i])
            self.scalers.append(Scaler)
        return X
    
    def transform(self, X):
        for i in range(X.shape[2]):
            X[:, :, i] = self.scalers[i].transform(X[:, :, i])
        return X
```

o zaman eğitim ve test verilerimizi dönüştürelim

```python
Feature_Scaler = MultidimensionScaler()
Xtrain = Feature_Scaler.fit_transform(Xtrain)
Xtest = Feature_Scaler.transform(Xtest)
```

```python
Target_Scaler = MinMaxScaler()
Ytrain = Target_Scaler.fit_transform(Ytrain.reshape(-1,1))
Ytest = Target_Scaler.transform(Ytest.reshape(-1,1))
```

```python
Xtrain.shape, Ytrain.shape, Xtest.shape, Ytest.shape
```

Daha sonra 2 fonksiyon oluşturuyoruz, nesneyi kaydediyoruz ve nesneyi yüklüyoruz.

```python
def save_object(obj, name : str):
    pickle_out= open(f"{name}.pck","wb")
    pickle.dump(obj, pickle_out)
    pickle_out.close()
    
def load_object(name : str):
    pickle_in = open(f"{name}.pck","rb")
    data = pickle.load(pickle_in)
    return data
```

```python
# nesnelerinizi gelecekteki amaçlar için kaydedin
save_object(Feature_Scaler, "Feature_Scaler")
save_object(Target_Scaler, "Target_Scaler")
```

# Model Oluşturma

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau

save_best = ModelCheckpoint("best_weights.h5", monitor='val_loss', save_best_only=True, save_weights_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.25, patience=5, min_lr=0.00001, verbose=1)
```

in this section, we use 'ReduceLROnPlateau' to reduce the learning rate when there is no improvement in the validation loss.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Bidirectional

model = Sequential()

model.add(Bidirectional(LSTM(512, return_sequences=True, recurrent_dropout=0.1, input_shape=(32,2))))
model.add(LSTM(256, recurrent_dropout=0.1))
model.add(Dropout(0.3))
model.add(Dense(64, activation='elu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='elu'))
model.add(Dense(1, activation='linear'))
```

burada aktivasyon olarak 'elu'yu kullanıyoruz. 'Elu', Relu ve Tanh aktivasyonunun birleştirilmiş hali gibi görünen, Leaky relu'nun değiştirilmiş bir versiyonudur.

```python
# optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
optimizer = tf.keras.optimizers.SGD(learning_rate=0.002)
model.compile(loss='mse', optimizer=optimizer)
```

SGD'yi optimize edici olarak kullanıyorum çünkü SGD, Adam'dan daha iyi bir performansa sahip.

```python
history = model.fit(Xtrain, Ytrain,
                   epochs=10,
                   batch_size=1,
                   verbose=1,
                   shuffle=False,
                   validation_data=(Xtest, Ytest),
                   callbacks=[reduce_lr, save_best])
```

```python
# en iyi ağırlıkları yükle

model.load_weights("best_weights.h5")
```

```python
Predictions = model.predict(Xtest)
```

Değerleri modelimize hedef olarak ölçeklendirdiğimiz için tahminler de ölçeklendirilecek

```python
Predictions = Target_Scaler.inverse_transform(Predictions)
Actual = Target_Scaler.inverse_transform(Ytest)
```

```python
Predictions.shape
```

```python
Predictions = np.squeeze(Predictions, axis=1)
Actual = np.squeeze(Actual, axis=1)
```

```python
# tahminleri ve gerçekleri kontrol edin
fig = go.Figure()

fig.add_trace(go.Scatter(x = data.index[-test_length:], y = Actual, mode ='lines', name='Gerçek Fiyat'))
fig.add_trace(go.Scatter(x = data.index[-test_length:], y = Predictions, mode = 'lines', name='Tahmin Edilmiş Fiyat'))
fig.show()
```
