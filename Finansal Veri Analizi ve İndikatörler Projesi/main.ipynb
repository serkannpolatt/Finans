```python
# RSI (Relative Strength Index) göstergesi hesaplama
dataframed['RSI'] = talib.RSI(dataframed['CLOSING PRICE'], timeperiod=14)
```

```python
# Stochastic Oscillator göstergesi hesaplama
slowk, slowd = talib.STOCH(dataframed['HIGHEST PRICE'], dataframed['LOWEST PRICE'], dataframed['CLOSING PRICE'])
dataframed['SlowK'] = slowk
dataframed['SlowD'] = slowd
```

```python
# Bollinger Bands göstergesi hesaplama
upper_band, middle_band, lower_band = talib.BBANDS(dataframed['CLOSING PRICE'], timeperiod=20)
dataframed['Upper_Band'] = upper_band
dataframed['Middle_Band'] = middle_band
dataframed['Lower_Band'] = lower_band
```

```python
# PPO göstergesi hesaplama
ppo = talib.PPO(dataframed['CLOSING PRICE'])
dataframed['PPO'] = ppo
```

```python
# Kısa vadeli ve uzun vadeli hareketli ortalamaları hesaplayın
short_term_ma = dataframed['TOTAL TRADED VOLUME'].rolling(window=12).mean()
long_term_ma = dataframed['TOTAL TRADED VOLUME'].rolling(window=26).mean()
```

```python
# PVO'yu hesaplayın
pvo = ((short_term_ma - long_term_ma) / long_term_ma) * 100
# PVO'yu veri çerçevesine ekleyin
dataframed['PVO'] = pvo
```

```python
# 10 günlük ROC hesaplama
roc_10 = talib.ROC(dataframed['CLOSING PRICE'], timeperiod=10)
dataframed['ROC_10'] = roc_10

# 60 günlük ROC hesaplama
roc_60 = talib.ROC(dataframed['CLOSING PRICE'], timeperiod=60)
dataframed['ROC_60'] = roc_60
```

```python
# RSI göstergesi hesaplama (genellikle 14 günlük periyotta hesaplanır)
rsi = talib.RSI(dataframed['CLOSING PRICE'], timeperiod=14)
dataframed['RSI'] = rsi

# Stoch RSI hesapla
fastk, fastd = talib.STOCHRSI(rsi, timeperiod=14, fastk_period=5, fastd_period=3)
```

```python
# Sonuçları veri çerçevesine ekleyin
dataframed['StochRSI_FastK'] = fastk
dataframed['StochRSI_FastD'] = fastd
```

```python
# 14 günlük Stochastic Oscillator hesapla
slowk, slowd = talib.STOCH(dataframed['HIGHEST PRICE'], dataframed['LOWEST PRICE'], dataframed['CLOSING PRICE'], fastk_period=14, slowk_period=3, slowd_period=3)

# Sonuçları veri çerçevesine ekleyin
dataframed['StochOscillator_SlowK'] = slowk
dataframed['StochOscillator_SlowD'] = slowd
```

```python
# TSI Indicator
# Kapanış fiyatlarının farkını hesaplayın
price_diff = dataframed['CLOSING PRICE'].diff()

# Kısa vadeli EMA (örneğin 13 günlük) hesapla
short_ema = talib.EMA(price_diff, timeperiod=13)

# Uzun vadeli EMA (örneğin 25 günlük) hesapla
long_ema = talib.EMA(price_diff, timeperiod=25)

# TSI hesapla
tsi = (short_ema - long_ema) / long_ema * 100

# Sonuçları veri çerçevesine ekleyin
dataframed['TSI'] = tsi
```

```python
# Ultimate Oscillator hesapla
# Periyotları belirleyin
period1 = 7
period2 = 14
period3 = 28

dataframed['LC'] = dataframed['CLOSING PRICE'].shift(1)
dataframed['BP'] = dataframed['CLOSING PRICE'] - dataframed['LC']
dataframed['TR'] = dataframed['HIGHEST PRICE'] - dataframed['LOWEST PRICE']

dataframed['UO'] = (
    4 * dataframed['BP'].rolling(window=period1).sum() / 
    dataframed['TR'].rolling(window=period1).sum() +
    2 * dataframed['BP'].rolling(window=period2).sum() / 
    dataframed['TR'].rolling(window=period2).sum() +
    dataframed['BP'].rolling(window=period3).sum() / 
    dataframed['TR'].rolling(window=period3).sum()
) * 100
```

```python
# Williams %R hesapla
# Periyodu belirleyin
period = 14

dataframed['H14'] = dataframed['HIGHEST PRICE'].rolling(window=period).max()
dataframed['L14'] = dataframed['LOWEST PRICE'].rolling(window=period).min()

dataframed['WilliamsR'] = ((dataframed['H14'] - dataframed['CLOSING PRICE']) / (dataframed['H14'] - dataframed['L14'])) * -100

# MACD Hattı hesapla
# 12 günlük ve 26 günlük EMA'ları hesapla
ema_12 = talib.EMA(dataframed['CLOSING PRICE'], timeperiod=12)
ema_26 = talib.EMA(dataframed['CLOSING PRICE'], timeperiod=26)

macd_line = ema_12 - ema_26

# Signal Hattı (9 günlük EMA) hesapla
signal_line = talib.EMA(macd_line, timeperiod=9)

# MACD Histogram hesapla
macd_histogram = macd_line - signal_line

# Sonuçları veri çerçevesine ekleyin
dataframed['MACD_Line'] = macd_line
dataframed['MACD_Signal'] = signal_line
dataframed['MACD_Histogram'] = macd_histogram

# EMV hesapla
dataframed['Midpoint'] = (dataframed['HIGHEST PRICE'] + dataframed['LOWEST PRICE']) / 2
dataframed['Box Ratio'] = (dataframed['HIGHEST PRICE'] - dataframed['LOWEST PRICE']) / dataframed['TOTAL TRADED VALUE']
dataframed['Multiplier'] = (dataframed['Midpoint'] - dataframed['Midpoint'].shift(1)) / (dataframed['HIGHEST PRICE'] - dataframed['LOWEST PRICE'])
dataframed['EMV'] = dataframed['Multiplier'] * dataframed['TOTAL TRADED VALUE']


# Force Index hesapla
dataframed['Force Index'] = (dataframed['CLOSING PRICE'] - dataframed['CLOSING PRICE'].shift(1)) * dataframed['TOTAL TRADED VOLUME']
```

```python
# MFI değerini hesaplayın
# Periyodu belirleyin
period = 14

# Money Flow değerini hesaplayın
dataframed['Money Flow'] = (dataframed['CLOSING PRICE'] - dataframed['CLOSING PRICE'].shift(1)) * dataframed['TOTAL TRADED VOLUME']

# Positive Money Flow ve Negative Money Flow değerlerini hesaplayın
dataframed['Positive Money Flow'] = dataframed['Money Flow'].apply(lambda x: x if x > 0 else 0)
dataframed['Negative Money Flow'] = dataframed['Money Flow'].apply(lambda x: -x if x < 0 else 0)

# Money Ratio değerini hesaplayın
dataframed['Money Ratio'] = dataframed['Positive Money Flow'].rolling(window=period).sum() / dataframed['Negative Money Flow'].rolling(window=period).sum()

dataframed['MFI'] = 100 - (100 / (1 + dataframed['Money Ratio']))
```

```python
# NVI hesapla
# Başlangıç değerini belirleyin
start_nvi = 1000

# NVI hesapla
nvi_values = [start_nvi]

for i in range(1, len(dataframed)):
    price_change = (dataframed.loc[i, 'CLOSING PRICE'] - dataframed.loc[i-1, 'CLOSING PRICE']) / dataframed.loc[i-1, 'CLOSING PRICE']
    volume_change = (dataframed.loc[i, 'TOTAL TRADED VOLUME'] - dataframed.loc[i-1, 'TOTAL TRADED VOLUME']) / dataframed.loc[i-1, 'TOTAL TRADED VOLUME']
    
    if volume_change > 0:
        nvi_values.append(nvi_values[-1] + (price_change * nvi_values[-1]))
    else:
        nvi_values.append(nvi_values[-1] + (price_change * nvi_values[-1]))

# NVI değerlerini veri çerçevesine ekleyin
dataframed['NVI'] = nvi_values
```

```python
#BB hesaplama
# Periyodu belirleyin
period = 20

# Hesaplama için boş bir liste oluşturun
upper_band = []
lower_band = []

for i in range(period, len(dataframed)):
    # Ortalama Hareket'i hesaplayın
    avg = dataframed.loc[i - period:i - 1, 'CLOSING PRICE'].mean()

    # Standart Sapma'yı hesaplayın
    std = dataframed.loc[i - period:i - 1, 'CLOSING PRICE'].std()

    # Üst Bant ve Alt Bant'ı hesaplayın
    upper = avg + (2 * std)
    lower = avg - (2 * std)

    upper_band.append(upper)
    lower_band.append(lower)

# Hesaplanan Üst Bant ve Alt Bant değerlerini veri çerçevesine ekleyin
dataframed['Upper Bollinger Band'] = [None] * period + upper_band
dataframed['Lower Bollinger Band'] = [None] * period + lower_band
```

```python
#DonchianChannel hesaplama
# Periyodu belirleyin
period = 20

# Hesaplama için boş bir liste oluşturun
upper_band = []
lower_band = []
middle_band = []

for i in range(period, len(dataframed)):
    # Son 20 gün içindeki en yüksek ve en düşük fiyatları bulun
    highest_high = dataframed.loc[i - period:i - 1, 'HIGHEST PRICE'].max()
    lowest_low = dataframed.loc[i - period:i - 1, 'LOWEST PRICE'].min()

    # Orta Bant, Üst Bant ve Alt Bant'ı hesaplayın
    middle = (highest_high + lowest_low) / 2
    upper = middle + (2 * (middle - lowest_low))
    lower = middle - (2 * (highest_high - middle))

    upper_band.append(upper)
    lower_band.append(lower)
    middle_band.append(middle)

# Hesaplanan Üst Bant, Alt Bant ve Orta Bant değerlerini veri çerçevesine ekleyin
dataframed['Upper Donchian Channel'] = [None] * period + upper_band
dataframed['Lower Donchian Channel'] = [None] * period + lower_band
dataframed['Middle Donchian Channel'] = [None] * period + middle_band
```

```python
#KeltnerChannel 
# Periyodu belirleyin
period = 20

# ATR (Average True Range) için periyodu belirleyin
atr_period = 14

# Katsayıyı belirleyin (genellikle 2 kullanılır)
coefficient = 2

# Hesaplama için boş bir liste oluşturun
upper_band = []
lower_band = []
middle_band = []

for i in range(period, len(dataframed)):
    # Son 20 günün kapanış fiyatlarını alın
    closing_prices = dataframed.loc[i - period:i - 1, 'CLOSING PRICE']

    # Orta Bant (Basit Hareketli Ortalama) hesaplayın
    middle = closing_prices.mean()

    # True Range (TR) hesaplayın
    true_ranges = []
    for j in range(1, atr_period + 1):
        highest_high = dataframed.loc[i - j, 'HIGHEST PRICE']
        lowest_low = dataframed.loc[i - j, 'LOWEST PRICE']
        previous_close = dataframed.loc[i - j - 1, 'CLOSING PRICE']
        true_range = max(highest_high - lowest_low, abs(highest_high - previous_close), abs(lowest_low - previous_close))
        true_ranges.append(true_range)

    # ATR (Average True Range) hesaplayın
    atr = sum(true_ranges) / atr_period

    # Üst Bant ve Alt Bant'ı hesaplayın
    upper = middle + (atr * coefficient)
    lower = middle - (atr * coefficient)

    upper_band.append(upper)
    lower_band.append(lower)
    middle_band.append(middle)

# Hesaplanan Üst Bant, Alt Bant ve Orta Bant değerlerini veri çerçevesine ekleyin
dataframed['Upper Keltner Channel'] = [None] * period + upper_band
dataframed['Lower Keltner Channel'] = [None] * period + lower_band
dataframed['Middle Keltner Channel'] = [None] * period + middle_band
```

```python
#UlcerIndex hesaplama
# Periyodu belirleyin
period = 14

# Hesaplama için boş bir liste oluşturun
ulcer_index = []

for i in range(period, len(dataframed)):
    # Periyodu içeren verileri alın
    closing_prices = dataframed.loc[i - period:i - 1, 'CLOSING PRICE']

    # En yüksek kapanış fiyatını bulun
    highest_close = closing_prices.max()

    # Bugünkü kapanış fiyatını alın
    today_close = dataframed.loc[i, 'CLOSING PRICE']

    # Düşüş yüzdesini hesaplayın
    percentage_drawdown = ((highest_close - today_close) / highest_close) * 100

    # Düşüş yüzdesini kareleyin ve listeye ekleyin
    ulcer_index.append(percentage_drawdown ** 2)

# Karelerin toplamını hesaplayın
sum_of_squares = sum(ulcer_index)

# Toplamı periyodun boyutuna bölerek Ulcer Index'i hesaplayın
ulcer_index_value = np.sqrt(sum_of_squares / period)

# Ulcer Index değerini veri çerçevesine ekleyin
dataframed['Ulcer Index'] = [None] * period + ulcer_index
```

```python
#Aroon Indicator hesaplama
# Periyodu belirleyin
period = 14

# Hesaplama için boş bir liste oluşturun
aroon_up = []
aroon_down = []

for i in range(period, len(dataframed)):
    # Periyodu içeren verileri alın
    high_prices = dataframed.loc[i - period:i - 1, 'HIGHEST PRICE']
    low_prices = dataframed.loc[i - period:i - 1, 'LOWEST PRICE']

    # En yüksek fiyatın hangi günlerde gerçekleştiğini sayın
    high_index = high_prices.idxmax()
    aroon_up_value = period - (period - (i - high_index + 1))

    # En düşük fiyatın hangi günlerde gerçekleştiğini sayın
    low_index = low_prices.idxmin()
    aroon_down_value = period - (period - (i - low_index + 1))

    aroon_up.append(aroon_up_value)
    aroon_down.append(aroon_down_value)

# Hesaplanan Aroon Up ve Aroon Down değerlerini veri çerçevesine ekleyin
dataframed['Aroon Up'] = [None] * period + aroon_up
dataframed['Aroon Down'] = [None] * period + aroon_down
```

```python
#CCI (Commodity Channel Index) Indicator hesaplama
# Dönemi belirleyin (örneğin, 14 günlük bir dönem)
period = 14

# Boş bir liste oluşturun
cci_values = []

# TP (Typical Price) sütununu oluşturun
dataframed['TP'] = (dataframed['HIGHEST PRICE'] + dataframed['LOWEST PRICE'] + dataframed['CLOSING PRICE']) / 3

# CCİ hesaplamalarını yapın
for i in range(period, len(dataframed)):
    # TP'nin ortalama değerini hesaplayın
    tp_mean = dataframed.loc[i - period:i - 1, 'TP'].mean()

    # Çalışma Aralığını hesaplayın
    mean_deviation = abs(dataframed.loc[i - period:i - 1, 'TP'] - tp_mean).mean()

    # CCI (Commodity Channel Index) değerini hesaplayın
    cci = (dataframed.loc[i, 'TP'] - tp_mean) / (0.015 * mean_deviation)

    # CCI değerini listeye ekleyin
    cci_values.append(cci)

# Boş değerler ekleyin (ilk dönem için)
cci_values = [None] * period + cci_values

# CCİ değerlerini veri çerçevesine ekleyin
dataframed['CCI'] = cci_values
```

```python
#KST (Know Sure Thing) Indicator hesaplama
# KST için kullanılacak periyotları belirleyin
short_roc_period = 10
medium_roc_period = 15
long_roc_period = 20
very_long_roc_period = 30
kst_period = 9

# ROC (Rate of Change) değerlerini hesaplayın
dataframed[f'ROC ({short_roc_period} GÜNLÜK)'] = dataframed['CLOSING PRICE'].pct_change(short_roc_period) * 100
dataframed[f'ROC ({medium_roc_period} GÜNLÜK)'] = dataframed['CLOSING PRICE'].pct_change(medium_roc_period) * 100
dataframed[f'ROC ({long_roc_period} GÜNLÜK)'] = dataframed['CLOSING PRICE'].pct_change(long_roc_period) * 100
dataframed[f'ROC ({very_long_roc_period} GÜNLÜK)'] = dataframed['CLOSING PRICE'].pct_change(very_long_roc_period) * 100

# KST hesaplaması için kullanılacak verileri hazırlayın
roc_sum = (
    dataframed[f'ROC ({short_roc_period} GÜNLÜK)'] +
    dataframed[f'ROC ({medium_roc_period} GÜNLÜK)'] +
    dataframed[f'ROC ({long_roc_period} GÜNLÜK)'] +
    dataframed[f'ROC ({very_long_roc_period} GÜNLÜK)']
)
kst_values = talib.SMA(talib.SMA(talib.SMA(talib.SMA(roc_sum, kst_period), kst_period), kst_period), kst_period)

# KST değerini veri çerçevesine ekleyin
dataframed['KST'] = [None] * (len(dataframed) - len(kst_values)) + list(kst_values)
```

```python
#Mass Index hesaplama
# Periyodu belirleyin
period = 9

# Hesaplama için boş bir liste oluşturun
mass_index_values = []

ema_multiplier = 2 / (period + 1)
ema_values = []

for i in range(1, len(dataframed)):
    # High-Low Range (HLR) hesaplayın
    high_low_range = dataframed.loc[i, 'HIGHEST PRICE'] - dataframed.loc[i, 'LOWEST PRICE']

    # Exponential Moving Average (EMA) of HLR hesaplayın
    if i == 1:
        ema = high_low_range
    else:
        ema = (high_low_range * ema_multiplier) + (ema_values[-1] * (1 - ema_multiplier))

    ema_values.append(ema)

    # EMA değerlerini toplayarak Mass Index'i hesaplayın
    if i >= period:
        mass_index = sum(ema_values[-period:])
        mass_index_values.append(mass_index)

# Hesaplanan Mass Index değerlerini veri çerçevesine ekleyin
dataframed['Mass Index'] = [None] * (len(dataframed) - len(mass_index_values)) + mass_index_values
```

```python
#STC (Stochastic Oscillator) Indicator hesaplama
# STC hesaplama periyodunu belirleyin
stc_period = 14

# %K değerlerini hesaplayın
dataframed['Lowest Low'] = dataframed['LOWEST PRICE'].rolling(window=stc_period).min()
dataframed['Highest High'] = dataframed['HIGHEST PRICE'].rolling(window=stc_period).max()
dataframed['%K'] = ((dataframed['CLOSING PRICE'] - dataframed['Lowest Low']) / (dataframed['Highest High'] - dataframed['Lowest Low'])) * 100

# %D değerlerini hesaplayın (3 günlük hareketli ortalama)
dataframed['%D'] = dataframed['%K'].rolling(window=3).mean()
```

```python
#TRIX (Triple Exponential Moving Average) Indicator hesaplama
# TRIX hesaplama periyodunu belirleyin
trix_period = 15

# İlk üssel hareketli ortalamayı hesaplayın
dataframed['EMA1'] = dataframed['CLOSING PRICE'].ewm(span=trix_period, adjust=False).mean()

# İkinci üssel hareketli ortalamayı hesaplayın
dataframed['EMA2'] = dataframed['EMA1'].ewm(span=trix_period, adjust=False).mean()

# TRIX değerini hesaplayın
dataframed['TRIX'] = (dataframed['EMA2'] - dataframed['EMA2'].shift(1)) / dataframed['EMA2'].shift(1) * 100
```

```python
#Vortex Indicator hesaplama
# Vortex Indicator hesaplama periyodunu belirleyin
vortex_period = 14

# True Range (TR) hesaplaması
dataframed['High-Low'] = dataframed['HIGHEST PRICE'] - dataframed['LOWEST PRICE']
dataframed['High-Close'] = abs(dataframed['HIGHEST PRICE'] - dataframed['CLOSING PRICE'].shift(1))
dataframed['Low-Close'] = abs(dataframed['LOWEST PRICE'] - dataframed['CLOSING PRICE'].shift(1))
dataframed['TR'] = dataframed[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)

# Vortex Positive (VI+) ve Vortex Negative (VI-) hesaplaması
dataframed['VM+'] = (dataframed['HIGHEST PRICE'].shift(1) - dataframed['LOWEST PRICE']) - (dataframed['HIGHEST PRICE'] - dataframed['CLOSING PRICE'].shift(1))
dataframed['VM-'] = (dataframed['LOWEST PRICE'].shift(1) - dataframed['HIGHEST PRICE']) - (dataframed['LOWEST PRICE'] - dataframed['CLOSING PRICE'].shift(1))
dataframed['VI+'] = dataframed['VM+'].rolling(window=vortex_period).sum()
dataframed['VI-'] = dataframed['VM-'].rolling(window=vortex_period).sum()

# Vortex Indicator (VI) hesaplaması
dataframed['Vortex Indicator'] = dataframed['VI+'] / dataframed['VI-']

result_df = dataframed[['TRADE DATE', 'INSTRUMENT SERIES CODE', 'INSTRUMENT NAME', 'MARKET SEGMENT', 'MARKET', 'INSTRUMENT TYPE',
                        'INSTRUMENT CLASS', 'MARKET MAKER', 'BIST 100 INDEX', 'BIST 30 INDEX', 'GROSS SETTLEMENT', 'SUSPENDED',
                        'OPENING PRICE', 'OPENING SESSION PRICE', 'LOWEST PRICE', 'HIGHEST PRICE', 'CLOSING PRICE',
                        'CLOSING SESSION PRICE', 'CHANGE TO PREVIOUS CLOSING (%)', 'REMAINING BID', 'REMAINING ASK', 'VWAP',
                        'TOTAL TRADED VALUE', 'TOTAL TRADED VOLUME', 'TOTAL NUMBER OF CONTRACTS', 'REFERENCE PRICE',
                        'TRADED VALUE AT OPENING SESSION', 'TRADED VOLUME AT OPENING SESSION',
                        'NUMBER OF CONTRACTS AT OPENING SESSION', 'TRADED VALUE AT CLOSING SESSION',
                        'TRADED VOLUME AT CLOSING SESSION', 'NUMBER OF CONTRACTS AT CLOSING SESSION',
                        'TRADED VALUE OF TRADES AT CLOSING PRICE', 'TRADED VOLUME OF TRADES AT CLOSING PRICE',
                        'NUMBER OF CONTRACTS OF TRADES AT CLOSING PRICE', 'LOWEST SHORT SALE PRICE',
                        'HIGHEST SHORT SALE PRICE', 'SHORT SALE VWAP', 'TRADED VALUE OF SHORT SALE TRADES',
                        'TRADED VOLUME OF SHORT SALE TRADES', 'NUMBER OF CONTRACTS OF SHORT SALE TRADES',
                        'LOWEST TRADE REPORT PRICE', 'HIGHEST TRADE REPORT PRICE', 'TRADE REPORT VWAP',
                        'TRADE REPORT TRADED VALUE', 'TRADE REPORT TRADED VOLUME', 'NUMBER OF TRADE REPORTS',
                        'RSI', 'SlowK', 'SlowD','PPO', 'PVO','ROC_10','ROC_60','StochOscillator_SlowK',
                        'StochOscillator_SlowD','TSI','UO','WilliamsR','MACD_Line','MACD_Signal','MACD_Histogram','EMV','Force Index',
                        'MFI','NVI','Upper Bollinger Band','Lower Bollinger Band','Upper Donchian Channel','Lower Donchian Channel',
                        'Middle Donchian Channel','Upper Keltner Channel','Lower Keltner Channel','Middle Keltner Channel','Ulcer Index',
                        'Aroon Up','Aroon Down','CCI','KST','Mass Index','%D','%K','TRIX','Vortex Indicator']]  

# result_df.to_csv('/kaggle/working/calculated_indicators_thyao.csv', index=False)
print("Indikatör hesaplama csv dosyası olusturuldu!!")
```

```python
# Günlük ortalama ve standart sapmayı hesaplayın
gunluk_ortalama = result_df['Gunluk_Degisim_Orani'].mean()
gunluk_standart_sapma = result_df['Gunluk_Degisim_Orani'].std()

# Etiket sütununu ekleyin
result_df['Etiket'] = 0  # Başlangıçta tüm değerlere "0" ata
result_df.loc[result_df['Gunluk_Degisim_Orani'] > (gunluk_ortalama + 2 * gunluk_standart_sapma), 'Etiket'] = 1  # 2 fazla ise 1 etiketi
result_df.loc[(result_df['Gunluk_Degisim_Orani'] == gunluk_ortalama - gunluk_standart_sapma) & (result_df['Gunluk_Degisim_Orani'] <= gunluk_ortalama + gunluk_standart_sapma), 'Etiket'] = 0  # Belirtilen aralıkta ise 0 etiketi
result_df.loc[result_df['Gunluk_Degisim_Orani'] < (gunluk_ortalama - 2 * gunluk_standart_sapma), 'Etiket'] = 2  # 2 eksik ise 2 etiketi

# Sonuçları yazdırın
print(result_df[['TRADE DATE', 'CLOSING PRICE', 'Gunluk_Degisim_Orani', 'Etiket']])
```
